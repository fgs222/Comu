# Ejercicio 1

Del esquema se obtienen las siguientes ecuaciones:

$$
V_{o1} = V_{i2}~~~~~~V_{o2} = V_{i3}~~~~~~V_{o3} = V_{oL}
$$

$$
Z_{o1} = Z_{i2}~~~~~~Z_{o2} = Z_{i3}~~~~~~Z_{o3} = Z_L
$$

Siendo la potencia de entrada de $A_1,~P_{i1}$:

$$
P_{i1} = \frac{V^2_{i1}}{Z_{i1}}~~\to~~V^2_{i1} = P_{i1}~.~Z_{i1}~~\to~~V_{i1} = \sqrt{P_{i1}~.~Z_{i1}}
$$

Remplazando por los valores de la tabla:

$$
V_{i1} = \sqrt{1x10^{-6}~W~.~50~\Omega}~~\to~~V_{i1} = 7,071x10^{-3}~V
$$

Siendo la ganancia en dB de $A_1,~G_1$:

$$
G_1 = 10~.~log(\frac{P_{o1}}{P_{i1}})~~\to~~log(\frac{P_{o1}}{P_{i1}}) = \frac{G_1}{10}~~\to~~\frac{P_{o1}}{P_{i1}} = 10^{\frac{G_1}{10}}
$$

$$
P_{o1} = P_{i1}~.~10^{\frac{G_1}{10}}
$$

Remplazando por los valores de la tabla:

$$
P_{o1} = 1x10^{-6}~W~.~10^{\frac{50~dB}{10}}~~\to~~P_{o1} = 0,1~W
$$

Siendo la ganancia en veces de $A_1,~G_1$:

$$
G_1 = \frac{P_{o1}}{P_{i1}}
$$

Remplazando por los valores de la tabla:

$$
G_1 = \frac{0,1~W}{1x10^{-6}~W}~~G_1 = 100000
$$

Siendo la potencia de salida de $A_1,~P_{o1}$:

$$
P_{o1} = \frac{V^2_{o1}}{Z_{o1}}~~\to~~V^2_{o1} = P_{o1}~.~Z_{o1}~~\to~~V_{o1} = \sqrt{P_{o1}~.~Z_{o1}}
$$

Siendo $V_{o1} = V_{i2}$ y remplazando por los valores de la tabla:

$$
V_{i2} = \sqrt{0,1~W~.~150~\Omega}~~\to~~V_{i2} = 3,873~V
$$

Siendo la ganancia en veces de $A_2,~G_2$:

$$
G_2 = \frac{P_{o2}}{P_{i2}}~~\to~~P_{o2} = G_2~.~P_{i2}
$$

Remplazando por los valores de la tabla:

$$
P_{o2} = 0,01~.~0,1~W~~\to~~P_{o2} = 1x10^{-3}~W
$$

Siendo la ganancia en dB de $A_2,~G_2$:

$$
G_2 = 10~.~log(\frac{P_{o2}}{P_{i2}})
$$

Remplazando por los valores de la tabla:

$$
G_2 = 10~.~log(\frac{1x10^{3}~W}{0,1~W})~~\to~~G_2 = -20~dB
$$

Siendo la potencia de salida de $A_2,~P_{o2}$:

$$
P_{o2} = \frac{V^2_{o2}}{Z_{o2}}~~\to~~V^2_{o2} = P_{o2}~.~Z_{o2}~~\to~~V_{o2} = \sqrt{P_{o2}~.~Z_{o2}}
$$

Siendo $V_{i3} = V_{o2}$ y remplazando por los valores de la tabla:

$$
V_{i3} = \sqrt{1x10^{-3}~W~.~600~\Omega}~~\to~~V_{i3} = 0,775~V
$$

Siendo la potencia de entrada de $A_2,~P_{i2}$:

$$
P_{o3} = \frac{V^2_{o3}}{Z_{o3}}
$$

Remplazando por los valores de la tabla:

$$
P_{o3} = \frac{(0,38~V)^2}{8~\Omega}~~\to~~P_{i3} = 18,05x10^{-3}~W
$$

Siendo la ganancia en veces de $A_3,~G_3$:

$$
G_3 = \frac{P_{o3}}{P_{i3}}
$$

Remplazando por los valores de la tabla:

$$
G_3 = \frac{18,05x10^{-3}~W}{1x10^{-3}~W}~~\to~~G_3 = 18,05
$$

Siendo la ganancia en dB de $A_3,~G_3$:

$$
G_3 = 10~.~log(\frac{P_{o3}}{P_{i3}})
$$

Remplazando por los valores de la tabla:

$$
G_3 = 10~.~log(\frac{18,05x10^{-3}~W}{1x10^{-3}~W})~~\to~~G_3 = 12,56~dB
$$

Siendo la potencia de entrada $P_{i1}$ en dBm:

$$
P_{i1} = 10~.~log(\frac{P_{i1}}{1~mW})
$$

Remplazando por los valores de la tabla:

$$
P_{i1} = 10~.~log(\frac{1x10^{-6}~W}{1~mW})~~\to~~P_{i1} = -30~dBm
$$

Siendo la potencia de entrada en $P_{i2}$ en dBm:

$$
P_{i2} = 10~.~log(\frac{P_{i2}}{1~mW})
$$

Remplazando por los valores de la tabla:

$$
P_{i2} = 10~.~log(\frac{0,1~W}{1~mW})~~\to~~P_{i1} = 20~dBm
$$

Siendo la potencia de entrada en $P_{i3}$ en dBm:

$$
P_{i3} = 10~.~log(\frac{P_{i3}}{1~mW})
$$

Remplazando por los valores de la tabla:

$$
P_{i2} = 10~.~log(\frac{1x10^{-3}~W}{1~mW})~~\to~~P_{i1} = 0~dBm
$$

Siendo la potencia de entrada en $P_{o3}$ en dBm:

$$
P_{o3} = 10~.~log(\frac{P_{o3}}{1~mW})
$$

Remplazando por los valores de la tabla:

$$
P_{o3} = 10~.~log(\frac{18,05x10^{-3}~W}{1~mW})~~\to~~P_{o3} = 12,56~dBm
$$

Siengo la ganancia $G_1$ en dBm:

$$
G_1 = P_{o1} - P_{i1}
$$

Remplazando por los valores de la tabla:

$$
G_1 = 20~dBm - (-30~dBm)~~\to~~G_1 = 50~dB
$$

Siengo la ganancia $G_2$ en dBm:

$$
G_2 = P_{o2} - P_{i2}
$$

Remplazando por los valores de la tabla:

$$
G_2 = 0~dBm - 20~dBm~~\to~~G_1 = -20~dB
$$

Siengo la ganancia $G_3$ en dBm:

$$
G_3 = P_{o3} - P_{i3}
$$

Remplazando por los valores de la tabla:

$$
G_3
= 12,56~dBm - 0~dBm~~\to~~G_1 = 12,56~dB
$$
