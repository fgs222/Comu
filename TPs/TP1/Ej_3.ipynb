### Parte a

Siendo la potencia de ruido expresada en dBm:

$$
P_{r~[dBm]} = 10~.~log(\frac{P_{r~[W]}}{1~mW})
$$

Remplazando por el valor del enunciado:

$$
P_{r~[dBm]} = 10~.~log(\frac{50~pW}{1~mW})~~\to~~P_{r~[dBm]} = -73,01~dBm
$$

Siendo la relacin seal ruido:

$$
snr_{[W]} = \frac{P_{s~[W]}}{P_{r~[W]}}~~\to~~P_{s~[W]} = snr_{[W]}~.~P_{r~[W]}~~\to~~P_{s~[W]} = 10^{\frac{snr{[dB]}}{10}}~.~P_{r~[W]}
$$

Remplazando por los valores del enunciado:

$$
P_{s~[W]} = 10^{\frac{30~dB}{10}}~.~50~pW~~\to~~P_{s~[W]} = 50~nW
$$

Que en dBm es:

$$
P_{s~[dBm]} = 10~.~log(\frac{50~nW}{1~mW})~~\to~~P_{s~[dBm]} = -43,01~dBm
$$

Se puede observar que el resultado de la potencia de seal en dBm puede calcularse como:

$$
P_{s~[dBm]} = snr_{[dB]} + P_{r~[dBm]}~~\to~~P_{s~[dBm]} = 30~dB - 73,01~dBm~~\to~~P_{s~[dBm]} = -43,01~dBm
$$

Siendo la ecuacin de la atenuacin del radioenlace:

$$
L_{bf} = 20~.~log(\frac{4 . \pi . d}{\lambda})~~~~~\lambda: longitud~de~onda~[m]
$$

$$
L_{bf} = 20~.~log(\frac{4 . \pi . d}{\frac{c}{f}})~~\to~~L_{bf} = 20~.~log(\frac{4 . \pi . d . f}{c})~~~~~c: velocidad~de~la~luz~[\frac{m}{s}]~~~~~~f: frecuencia~[Hz]
$$

Siendo la potencia de transmicion:

### Parte c

Siendo la atenuacion del radioenlace para los 3 casos:

$$
L_{bf}(5km) = 98,46~dB~~~~~~L_{bf}(10km) = 104,48~dB~~~~~~L_{bf}(20km) = 110,5~dB
$$

Se puede observar que la diferencia en dB en cada aumento de atenuacion es de 6 dB (la atenuacion se cuatriplica). 

Siendo la atenuacion del coaxil para los 3 casos:

$$
A_{t~(5~km)} = 30~dB~~~~~~A_{t~(10~km)} = 60~dB~~~~~~A_{t~(20~km)} = 120~dB
$$

- Se oberserva que el radioenlace comienza con mayor atenuacion pero aumenta en menor medida a medida que se aumenta la distancia.
- El enlace coaxil con menor atenuacion pero aumenta considerablemente a medida que aumenta la distancia.
- Es preferible utilizar coaxil para distancias cortas y radioenlace para distancias largas.
