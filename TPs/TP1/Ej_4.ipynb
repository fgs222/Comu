### Parte A

$$
x(t) = \sum_{-\infty}^{+\infty} C_n~.~e^{-jn\omega_0t}~~~~~~siendo~~~C_n = \frac{1}{T_0}~\int_{-\frac{\tau}{2}}^{\frac{\tau}{2}} A~.~e^{-jn\omega_0t}~dt
$$

$$
C_n = \frac{1}{T}~\int_{-\frac{\tau}{2}}^{\frac{\tau}{2}} A~.~e^{-jn\omega_0t}~dt = \frac{A}{T}~.~[\frac{1}{-jn\omega_0}~.~e^{-jn\omega_0t}]\Bigl|_{-\frac{\tau}{2}}^{\frac{\tau}{2}}
$$

$$
C_n = \frac{A}{T}~.~\frac{1}{jn\omega_0}~.~[e^{\frac{-jn\omega_0\tau}{2}} - e^{\frac{jn\omega_0\tau}{2}}]~~\to~~C_n = \frac{A}{T}~.~\frac{1}{jn\omega_0}~.~2j~.~\frac{e^{\frac{-jn\omega_0\tau}{2}} - e^{\frac{jn\omega_0\tau}{2}}}{2j}
$$

$$
C_n = \frac{2A}{Tn\omega_0}~.~sen(\frac{n\omega_0\tau}{2})~~~~~~siendo~~\omega_0 = 2\pi f = \frac{2\pi}{T}
$$

$$
C_n = \frac{2AT}{Tn2\pi}~.~sen(\frac{n2\pi\tau}{2T})~~\to~~C_n = \frac{A}{n\pi}~.~sen(\frac{n\pi\tau}{T})
$$

$$
C_n = \frac{A}{n\pi}~.~\frac{n\pi\tau}{T}~.~\frac{sen(\frac{n\pi\tau}{T})}{\frac{n\pi\tau}{T}}~~\to~~C_n = \frac{A\tau}{T}~.~sinc(\frac{n\pi\tau}{T})
$$

### Parte C

$$

$$

Potencia normalizada total en el dominio del tiempo

$$
P = \frac{1}{T}~\int_{0}^T |x(t)|^2~dt = \frac{1}{T}~\int_{0}^\tau A^2~dt = \frac{A^2}{T}~[t]\Bigl|_{0}^{\tau} = \frac{A^2}{T}~.~(\tau - 0)
$$

$$
P = \frac{1}{100~ms}~.~(20 ms)~~\to~~P = \frac{1}{5}
