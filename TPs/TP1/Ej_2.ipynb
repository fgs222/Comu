# Ejericio 2

### Parte a

Siendo la potencia de seal del enlace 0 en dBW:

$$
P_{s0~[dBW]} = 10~.~log(\frac{P_{s0~[W]}}{1~W})~~\to~~\frac{P_{s0~[dBW]}}{10} = log(\frac{P_{s0~[W]}}{1~W})~~\to~~\frac{P_{s0~[W]}}{1~W} = 10^{\frac{P_{s0~[dBW]}}{10}}
$$

$$
P_{s0~[W]} = 1~W~.~10^{\frac{P_{s0~[dBW]}}{10}}
$$

Remplazando por los valores del enunciado:

$$
P_{s0~[W]} = 1~mW . 10^{\frac{10~dBW}{10}}~~\to~~P_{s0~[W]} = 10~W
$$

Siendo la atenuacion del enlace 1 en dB:

$$
At_{1~[dB]} = 10~.~log(\frac{P_{s1~[W]}}{P_{s0~[W]}})~~\to~~log(\frac{P_{s1~[W]}}{P_{s0~[W]}}) = \frac{At_{1~[dB]}}{10}~~\to~~\frac{P_{s1~[W]}}{P_{s0~[W]}} = 10^{\frac{At_{1~[dB]}}{10}}
$$

$$
P_{s1~[W]} = P_{s0~[W]}~.~10^{\frac{At_{1~[dB]}}{10}}
$$

Remplazando por los valores del enunciado:

$$
P_{s1~[W]} = 10~W~.~10^{\frac{-40~dB}{10}}~~\to~~P_{s1~[W]} = 1x10^{-3}~W
$$

Siendo la potencia de ruido del enlace 1 en dBm:

$$
P_{n1~[dBm]} = 10~.~log(\frac{P_{n1~[W]}}{1~mW})~~\to~~\frac{P_{n1~[dBm]}}{10} = log(\frac{P_{n1~[W]}}{1~mW})~~\to~~\frac{P_{n1~[W]}}{1~mW} = 10^{\frac{P_{n1~[dBm]}}{10}}
$$

$$
P_{n1~[W]} = 1~mW~.~10^{\frac{P_{n1~[dBm]}}{10}}
$$

Remplazando por los valores del enunciado:

$$
P_{n1~[W]} = 1~mW~.~10^{\frac{-40~dBm}{10}}~~\to~~P_{n1~[W]} = 10x10^{-6}~W
$$

Siendo la relacion seal ruido en el enlace 1 en dB:

$$
snr_{[db]} = 10~.~log(\frac{P_{s1~[W]}}{P_{n1~[W]}})
$$

Remplazando por los valores calculados previamente:

$$
snr_{1~[db]} = 10~.~log(\frac{1x10^{-3}~W}{10x10^{-6}~W})~~\to~~snr_{[db]} = 20~dB
$$

### Parte b

Siendo la atenuacion del enlace 2 en dB:

$$
At_{2~[dB]} = 10~.~log(\frac{P_{s2~[W]}}{P_{s1~[W]}})~~\to~~log(\frac{P_{s2~[W]}}{P_{s1~[W]}}) = \frac{At_{2~[dB]}}{10}~~\to~~\frac{P_{s2~[W]}}{P_{s1~[W]}} = 10^{\frac{At_{2~[dB]}}{10}}
$$

$$
P_{s2~[W]} = P_{s1~[W]}~.~10^{\frac{At_{2~[dB]}}{10}}
$$

Remplazando por los valores previamente calculados:

$$
P_{s2~[W]} = 1x10^{-3}~W~.~10^{\frac{-12~dB}{10}}~~\to~~P_{s2~[W]} = 63,1x10^{-6}~W
$$

Calculando la atenuacion del ruido en el enlace 2:

$$
P'_{n2~[W]} = P_{n1~[W]}~.~10^{\frac{At_{2~[dB]}}{10}}
$$

Remplazando por los valores previamente calculados:

$$
P'_{n2~[W]} = 10x10^{-6}~.~10^{\frac{-12}{10}}~~\to~~P'_{n2~[W]} = 0,631x10^{-6}~W
$$

Siendo que el ruido empeora 16 veces en el enlace 2:

$$
P_{n2~[W]} = 16~.~P'_{n2~[W]}
$$

Remplazando por el valor calculado anteriormente:

$$
P_{n2~[W]} = 16~.~0,631x10^{-6}~W~~\to~~P_{n2~[W]} = 10,1x10^{-6}~W
$$

Siendo la relacion seal ruido en el enlace 2 en dB:

$$
snr_{2~[db]} = 10~.~log(\frac{P_{s2~[W]}}{P_{n2~[W]}})
$$

Remplazando por los valores calculados anteriormente:

$$
snr_{2~[db]} = 10~.~log(\frac{63,1x10^{-6}~W}{10,1x10^{-6}~W})~~\to~~snr_{2~[db]} = 7,96~dB
$$

### Parte c

Si aumentamos $P_{s0}$ para que $snr_2$ gane 6 dB:

$$
snr_{2~[dB]} = 13,96~dB
$$

Siendo $snr_2$:

$$
snr_{2~[db]} = 10~.~log(\frac{P_{s2~[W]}}{P_{n2~[W]}})~~\to~~log(\frac{P_{s2~[W]}}{P_{n2~[W]}}) = \frac{snr_{2~[db]}}{10}~~\to~~\frac{P_{s2~[W]}}{P_{n2~[W]}} = 10^{\frac{snr_{2~[db]}}{10}}
$$

$$
P_{s2~[W]} = P_{n2~[W]}~.~10^{\frac{snr_{2~[db]}}{10}}
$$

Remplazando por los valores previamente calculados:

$$
P_{s2~[W]} = 10,1x10^{-6}~.~10^{\frac{13,96~dB}{10}}~~\to~~P_{s2~[W]} = 251,4x10^{-6}~W
$$

De la ecuacion de la atenuacion en el enalce 2 obtenemos:

$$
At_{2~[dB]} = 10~.~log(\frac{P_{s2~[W]}}{P_{s1~[W]}})~~\to~~P_{s1~[W]} = \frac{P_{s2~[W]}}{10^{\frac{At_{2~[dB]}}{10}}}
$$

Remplazando por los valores previamente calculados:

$$
P_{s1~[W]} = \frac{251,4x10^{-6}~W}{10^{\frac{-12~dB}{10}}}~~\to~~P_{s1~[W]} = 3,98x10^{-3}~W
$$

De la ecuacion de la atenuacion en el enlace 1 obtenemos:

$$
At_{1~[dB]} = 10~.~log(\frac{P_{s1~[W]}}{P_{s0~[W]}})~~\to~~P_{s0~[W]} = \frac{P_{s1~[W]}}{10^{\frac{At_{1~[dB]}}{10}}}
$$

Remplazando por los valores previamente calculados:

$$
P_{s0~[W]} = \frac{3,98x10^{-3}~W}{10^{\frac{-40~dB}{10}}}~~\to~~P_{s0~[W]} = 39,8~W
$$

### Parte d

Considerando que el valor de la relacion seal ruido es dependiente del cociente entre la potencia de la seal y la potencia del ruido, y siendo que la atenuacin afecta a ambos fenmenos por igual, la relacin no se veria afectada por un aumento de atenuacin en el enlace.
