Podemos considerar que la funcion $f(t)$ esta compuesta por el producto de dos funciones $f(t) = f1(t)~.~f2(t)~$ <br>
Siendo $f1(t)$ una funcion senoidal con $f_c = 10~MHz$ y $f2(t)$ un tren de pulsos rectangulares con duty activo $\tau_{activo} = 1 \mu s$ y reposo $\tau_{reposo} = 1 \mu s$.

$$
f_1(t) = \left\{
    \begin{array}\\
        A_1 & \mbox{si } \ 0 < t < T/2  \\
        0 & \mbox{si } \ T/2 < t < T \\
    \end{array}
\right.
$$

$$
f_2 = A_2~.~cos(\omega_ct)
$$

Siendo:

$$
f(t) = f_1(t)~.~f_2(t)~~\to~~F(\omega) = F_1(\omega) * F_2(\omega)
$$

- Analizando $f_1$:

Recordando la serie trigonometrica de fourier

$$
f(t) = \frac{a_0}{2}~\sum_{1}^{\infty} C_n~.~sen(n \omega_0 t + \theta_t)~~~~~~~~~~~con~C_n = \sqrt{a^2_n + b^2_n}
$$

Por ser una seal impar:

$$
a_n = 0
$$

$$
a_n = \frac{2}{T}~\int_{0}^{T} f(t)~.~cos(n \omega_0 t)~dt~~\to~~a_0 = \frac{2}{T}~\int_{0}^{T} f(t)~dt = \frac{2}{T}~.~A_1~.~\frac{T}{2}~~\to~~a_0 = A_1~~\to~~\frac{a_0}{2} = \frac{A_1}{2}
$$

$$
b_n = \frac{2}{T}~\int_{0}^{T} f(t)~.~sen(n \omega_0 t)~dt = \frac{2~A_1}{T}~\int_{0}^{T} sen(n \omega_0 t)~dt = \frac{2~A_1}{T}~.~(\frac{-cos(n \omega_0 t)}{n \omega_0})\Bigl|_{0}^{\frac{T}{2}}
$$
