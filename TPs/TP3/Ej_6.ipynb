# Ejercicio 6

Siendo la seal de NBFM:

$$
S(t) = A_c~.~(cos(\omega_c~t) - A_c~.~\Phi_t~.~sen(\omega_c~t))
$$

Siendo la potencia de portadora:

$$
P_c = -43~dBm = 10~.~log(\frac{x}{1~mW})~~\to~~\frac{x}{1~mW} = 10^{\frac{-43~dBm}{10}}~~\to~~x = 1~mW~.~10^{\frac{-43~dBm}{10}}
$$

$$
P_c = 50,12~nW
$$

Siendo la expresion de la potencia de la portadora:

$$
P_c = \frac{A^2_c}{2Z}~~\to~~A_c = \sqrt{2P_c~.~Z} = \sqrt{2~.~50,12~nW~.~50~\Omega}
$$

$$
A_c = 2,24~mV
$$

### Parte A

Siendo la amplitud del mensaje:

$$
A_m = \frac{A_c~.~\beta}{2} = \frac{2,24~mV~.~0,157}{2}~~\to~~A_m = 175,84~\mu V
$$

Siendo la potencia de una banda lateral:

$$
P_m = \frac{A^2_m}{4Z} = \frac{(175,84 \mu V)^2}{4~.~50~\Omega}~~\to~~P_m = 154,6~pW
$$

### Parte B

Siendo la potencia de la banda lateral en AM y despejando el indice de modulacion:

$$
P_m = \frac{m^2~.~P_c}{4}~~\to~~m = \sqrt{\frac{4P_m}{P_c}} = \sqrt{\frac{4~.~154,6~pW}{50,12~nW}}
$$

$$
m = 0,111
$$

### Parte C

Siendo el indice de modulacion de AM:

$$
m = \frac{A_m}{A_c} = \frac{175,84~\mu V}{2,24~mV}~~\to~~m = 0,0785
$$

### Parte D

Siendo la seal de NBFM:

$$
S(t) = A_c~.~(cos(\omega_c~t) - A_c~.~\Phi_t~.~sen(\omega_c~t))
$$

Limitando la seal:

$$
S(t) = V_L~.~(cos(\omega_c~t) - V_L~.~\Phi_t~.~sen(\omega_c~t))
$$

Diferenciado la seal:

$$
S'(t) = -V_L~.~sen(\omega_c~t) - V_L~.~\Phi_t~.~cos(\omega_c~t) = -V_L~.~sen(\omega_c~t)~.~(1 + \Phi_t~.~cos(\omega_c~t))
$$

Debido a que la salida del detector de envolvente va a ser proporcional a la seal derivada:

$$
S_o(t) = |-V_L~.~(1 + cos(\omega_c~t))|
$$

Filtrando la corriente continua:

$$
S_o(t) = V_L~.~cos(\omega_c~t)
$$

Del analisis anterior se observa que el tono que se escucharia a la salida del receptor seria de la frecuencia de la portadora.

### Parte E

Siendo la seal de NBFM:

$$
S(t) = A_c~.~(cos(\omega_c~t) - A_c~.~\Phi_t~.~sen(\omega_c~t)) = A_c(cos(\omega_c~t) - \Phi_t~.~sen(\omega_c~t))
$$

$$
S(t) = A_c~.~cos(\omega_c~t)~.~(1+\Phi_t~.~tan(\omega_c~t))
$$

Siendo la seal de AM:

$$
S(t) = A_c~.~cos(\omega_c~t)~.~(1 + m(t))
$$

Por lo tanto el mensaje que se recibira seria:

$$
m(t) = \Phi_t~.~tan(\omega_c~t)
$$
