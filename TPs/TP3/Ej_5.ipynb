### Parte I

Siendo:

$$
3~dB = 10~.~log(x)~~\to~~x = 10^{\frac{3~dB}{10}} = 1,995~~veces
$$

Siendo la potencia normalizada:

$$
P_n = \frac{A^2_c}{2}
$$

$$
A_c = \sqrt{2~.~P_{ot}} = \sqrt{2~.~2~kW}~~\to~~A_c = 20\sqrt{10}
$$

Siendo la potencia de la componente espectral a frecuencia portadora:

$$
P'_n = \frac{2~kW}{1,995}~~\to~~P'_n = 1003~W
$$

Donde:

$$
P'_n = \frac{A^2_c}{2}~.~\sum_{n = -\infty}^{+\infty}~J^2_N(\beta)
$$

Para la componente de frecuencia de portadora:

$$
P'_0 = \frac{A^2_c}{2}~.~J^2_0(\beta)~~\to~~J^2_0(\beta) = P'_0~.~\frac{2}{A^2_c}
$$

$$
J_0(\beta) = \sqrt{1003~W~.~\frac{2}{(20\sqrt{10})^2}} = 0,708
$$

Del grafico, obtenemos que para un valor de 0,708 en la componente de portadora se obtiene un $\beta$ = 1,2
