# Ejercicio 1

### Parte A

Siendo el ancho de banda de FM por regla de Carson:

$$
BW(FM) = 2~.~[\Delta f + f_m]~~~~~~~~siendo~~~~\Delta f = K_f~.~A_m
$$

$$
\Delta f = 1~\frac{kHz}{V}~.~1~V~~\to~~\Delta f = 1~kHz
$$

$$
BW(FM) = 2~.~[1~kHz + 10~kHz]~~\to~~BW(FM) = 22~kHz
$$

Por lo tanto se demuestra que el ancho de banda ocupado por la seal de FM es mayor a 2 kHz.

### Parte B

$$
\Delta f = 1~\frac{MHz}{V}~.~1~V~~\to~~\Delta f = 1~MHz
$$

$$
BW(FM) = 2~.~[1~MHz + 10~kHz]~~\to~~BW(FM) = 2,02~MHz
$$

La diferencia respecto del punto a se debe a que $\Delta f >> f_m$. Por lo tanto es posible aproximar la regla de Carson para calcular el ancho de banda de la seal a:

$$
BW(FM) = 2~.~\Delta f
$$

### Parte C

- Parte a: $~~~BW(FM) = 22~kHz$
- Parte b: $~~~BW(FM) = 2,02~MHz$
