### Parte B

Siendo:

$$
\Delta f = k_f~.~A_m
$$

Siendo que:

$$
m_{AM} = m_{FM}~~\to~~\frac{A_{max} - A_{min}}{A_{max} + A_{min}} = \frac{\Delta f}{f_m}
$$

Por lo tanto:

$$
\Delta f = f_m~.~\frac{A_{max} - A_{min}}{A_{max} + A_{min}} = \frac{81640}{2\pi}~.~\frac{11 - 9}{11 + 9}
$$

$$
\Delta f = 1299,34 Hz
$$

### Parte C

Sabiendo que:

$$
\left\{
    \begin{array}\\
        A_{max} = A_c + A_m \\
        A_{min} = A_c - A_m 
    \end{array}
\right.
$$

Resolviendo el sistema de ecuaciones se obtiene que:

$$
A_c = 10~V~~~~~~~~A_m = 1~V
$$

Obteniendo la constante de modulacion en frecuencia:

$$
k_f = \frac{\Delta f}{A_m}~~\to~~k_f = 1299,34~\frac{Hz}{V}
$$
