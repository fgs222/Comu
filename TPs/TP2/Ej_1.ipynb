\left\{
    \begin{array}\\
        A_{max} = A_c + A_m = A_c~.~(1 + m)~~;~~m = \frac{A_m}{A_c} \\
        A_{min} = A_c - A_m \\
    \end{array}
\right.
$$

$$
5~V = A_c - A_m~~\to~~A_m = A_c - 5~V

### Parte C

$$
P_t = P_c + P_{BLI} + P_{BLS} = P_C + 2~.~P_{BL}
$$

$$
P_c = \frac{A^2_c}{2R} = \frac{(20~V)^2}{2~.~50~\Omega}~~\to~~P_c = 4~W
$$

$$
P_{BL} = \frac{(\frac{A_c~.~m}{2})^2}{2R} = \frac{(\frac{20~V~.~0,75}{2})^2}{2~.~50~\Omega}~~\to~~P_{BL} = 0,075~W
$$

$$
P_{t~[W]} = 4~W + 2~.~0,075~W~~\to~~P_{t~[W]} = 4,15~W
$$

$$
P_{t~[dB]} = 20~.~log(4,15~W)~~\to~~P_{t~[dB]} = 12,36~dB
$$

$$
P_{t~[dBm]} = 10~.~log(\frac{4,15~W}{1~mW})~~\to~~P_{t~[dBm]} = 36,18~dBm
$$
