### Parte A

Siendo la distancia que hay en banda base entre ambas bandas:



$$
\Delta_f = 2~.~f_i = 2~.~300~Hz~~\to~~\Delta_f = 600~Hz
$$

Para poder obtener una seal USB con el filtro especificado, es necesario que se cumpla la siguiente condicion:

$$
\Delta_f \geq (f_0 - \frac{B_t}{2}) - 0,98~.~(f_0 - \frac{B_t}{2})
$$

$$
\Delta_f \geq 0,02~.~(f_0 - \frac{B_t}{2})
$$

Siendo $B_t$:

$$
B_t = f_s - f_i = 3400~Hz - 300~Hz~~\to~~B_t = 3100~Hz
$$

Por lo tanto:

$$
600~Hz \geq 0,02~.~(26000~kHz - \frac{3100~Hz}{2})~~\to~~600~Hz \geq 519969~Hz
$$

Se observa que no se cumple el requerimiento, por lo tanto esta solucion no es posible.

### Parte C

Es necesario encontrar una frecuencia central del filtro que cumpla la siguiente condicion:

$$
\frac{\Delta_f}{0,02} + \frac{B_t}{2} \geq f_{01}
$$

Remplazando por los valores calculados anteriormente:

$$
\frac{600~Hz}{0,02} + \frac{3100~Hz}{2} \geq f_{01}~~\to~~31550~Hz \geq f_{01}
$$

Se observa que la frecuencia central del filtro puede tomar valores hasta un maximo de 31550 Hz.

Se propone:

$$
f_{01} = 30~kHz
$$

Por lo tanto, la frecuencia del oscilador del detector de producto es:

$$
f_{osc1} = f_{01} - (\frac{B_t}{2} + \Delta_f) = 30~kHz - (\frac{3100~Hz}{2} + 600~Hz)
$$

$$
f_{osc1} = 28150~Hz
$$

Una vez eliminada la banda inferior, es necesario llevar la seal USB a 26000 kHz. Para esto es necesario llevar la seal por un nuevo detector de producto donde:

$$
\frac{\Delta_{f2}}{0,02} + \frac{B_t}{2} \geq f_{02}
$$

Con:

$$
\Delta f_2 = 2~.~(f_{osc1} + 300~Hz)
$$

$$
\Delta f_2 = 2~.~(28150~Hz + 300~Hz)~~\to~~\Delta f_2 = 56900~Hz
$$

Remplazando en la inecuacion anterior:

$$
\frac{56900~Hz}{0,02} + \frac{3100~Hz}{2} \geq f_{02}~~\to~~2846,55 kHz \geq f_{02}
$$

Se observa que la frecuencia central del filtro puede tomar valores hasta un maximo de 2846,55 kHz.

Se propone:

$$
f_{02} = 2800~kHz
$$

Por lo tanto, la frecuencia del oscilador del detector de producto es:

$$
f_{osc2} = f_{02} - (\frac{B_t}{2} + \Delta_{f2}) = 2800~kHz - (\frac{3100~Hz}{2} + 56900~Hz)
$$

$$
f_{osc2} = 2741,55~kHz
$$

Se puede observar que es necesario implementar una 3era etapa, por lo tanto:

$$
\frac{\Delta_{f3}}{0,02} + \frac{B_t}{2} \geq f_{03}
$$

Con:

$$
\Delta f_3 = 2~.~(f_{osc2} + 300~Hz)
$$

$$
\Delta f_3 = 2~.~(2741,55~kHz + 300~Hz)~~\to~~\Delta f_3 = 5483,7~kHz
$$

Remplazando en la inecuacion anterior:

$$
\frac{5483,7~kHz}{0,02} + \frac{3100~Hz}{2} \geq f_{03}~~\to~~274186,55~kHz \geq f_{03}
$$

Por lo tanto, la frecuencia del oscilador del detector de producto es, con una frecuencia central de 26000~kHz:

$$
f_{osc3} = f_{03} - (\frac{B_t}{2} + \Delta_{f3}) = 26000~kHz - (\frac{3100~Hz}{2} + 5483,7~kHz)
$$

$$
f_{osc3} = 20514,75~kHz
$$
