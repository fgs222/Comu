# Ejercicio 3

### Parte A

Filtro de RF:

- Su funcon es filtrar la frecuencia imagen y todo lo que este despues de ella.
- Se encuentra vinculado al oscilador local ya que su filtrado depende de la posicion de dicho oscilador.
- Es un filtro pasabanda

Amplificador RF:

- Amplifica todo el espectro de seales recibidas y que pasaron el filtro RF.
- Su objetivo es llevar el nivel de las seales recibidas al requerido por el mezclador.

Oscilador local:

- Es un oscilador, vinculado al filtro de RF, cuyo objetivo es generar una seal que, mezclada con la seal recibida por la antena, obtenga como resultado una frecuencia deseada pero ubicada en una posicion fija en el espectro denominada "Frecuencia Intermedia".
- La frecuencia de este oscilador es veriable de forma tal de poder recorrer toda la banda del tipo de seal que se este trabajando (en el caso de AM desde 535 kHz a 1605 kHz)

Mezclador:

- Es el encargado de hacer la convolucion en frecuencia entre la seal recibida por antena y el oscilador local, centrada a la seal de interes en la denominada frecuencia intermedia.
- Se puede trabajar tanto con la componente "$A_x$ + fase" como con la componente "$A_x$ - fase" entre ambas seales, aunque se prefiere la opcion de oscilador por encima por ventajas constructivas del oscilador.

Filtro de FI:

- Es un filtro pasabanda de buena calidad.
- Son fijos, ya que en esta etapa ya se realizo el desplazamiento a FI.
- Su funcion es filtrar los canales adyacentes.

Amplificador de FI:

- Su objetivo es llevar la seal posterior al filtro de FI al nivel necesario para poder entrar al detector.

Detector:

- El objetivo es obtener la seal modulante
- Existen varios tipos, dependiendo de la modulacion empleada y las caracteristicas del trasmisor y receptor.
- Puede ser detector de envolvente, de producto, etc.
- Es la etapa donde se encuentra el control automatico de ganancia, el cual ajusta la seal recuperada en funcion del nivel de portadora recibida.

Control de volumen:

- Ajusta el nivel de seal a la salida del detector antes de ser amplificado una vez mas hacia el parlante de salida.

Amplificador de audio:

- Ultimo amplificador del conjunto.
- Ajusta la seal y entrega la potencia necesaria para poder generar la onda audible por el ser humano.

### Parte B

Siendo la potencia recibida en la antena:

$$
P_{Rx} = P_c + P_{BLS} + P_{BLI}
$$

Donde la potencia de la portadora es:

$$
P_c = \frac{A^2_c}{2Z}~~~~~~~~con~~~A_c = \sqrt{2}~.~V_{CEF}~~~~~~~~Z = 50~\Omega
$$

$$
A_c = \sqrt{2}~.~36~\mu V = 50,91~\mu V
$$

$$
P_c = \frac{(50,91~\mu V)^2}{2~50~\Omega}~~\to~~P_c = 25,92~pW
$$

Siendo la potencia de la banda lateral:

$$
P_{BL} = \frac{(A_c~.~m~.~0,5)^2}{2~.~Z} = \frac{(50,91~\mu V~.~0.8~.~0,5)^2}{2~.~50~\Omega}~~\to~~P_{BL} = 4,14~pW
$$

$$
P_{Rx}~[W] = 25,92~pW + 2~.~4,14~pW~~\to~~P_{Rx}~[W] = 34,21~pW
$$

$$
P_{Rx}~[dBm] = 10~log(\frac{P_{Rx}~[W]}{1~mW})~~\to~~ P_{Rx}~[dBm] = -74,56~dBm
$$

Siendo la ganancia total del sistema:

$$
G_{T} = G_{RF} + G_{MIX} + G_{FI} + G_{DET} + G_{OUT}~~\to~~G_{T} = 96~dB
$$

Siendo la potencia de salida del sistema:

$$
P_{par} = P_{Rx} + G_{T} = - 74,56~dBm + 96~dB~~\to~~P_{par} = 21,33~dBm
$$

### Parte C

Se observa una frecuencia imagen interferencia que se puede calcular de la siguiente manera:

$$
F_{imag} = F_s + 2~.~F_{i} = 690~kHz + 2~.~455~kHz~~\to~~F_{imag} = 1600~kHz
$$

- La frecuencia imagen siempre se encontrara a 2 $F_i$ de la frecuencia de interes, porque es el doble de la distancia entre esta frecuencia y la del oscilador.
- Esta interferencia no deberia ocurrir si el filtro de RF funcionara correctamente.

### Parte D

Para sintetizar 710 kHz necesito sintonizar el oscilador en la siguiente frecuencia:

$$
f_{osc} = 710~kHz + 455~kHz~~\to~~f_{osc} = 1165~kHz
$$

Si el oscilador, por alguna razon, generara frecuencias dinamicas que son multiplos de la frecuencia sintonizada, entonces:

$$
2~.~f_{osc} = 2330~kHz~~~~~~~~~~~~~3~.~f_{osc} = 3495~kHz
$$

Por lo tanto no seria posible sintonizar las frecuencias del enunciado a menos que se presente alguna alinealidad en el mezclador.
