# Ejercicio 2

### Parte A

Siendo la potencia normalizada de la portadora en Watts:

$$
P_c~[W] = \frac{A^2_c}{2} = \frac{(150~V)^2}{2}~~\to~~P_c~[W] = 11250~W
$$

Siendo la potencia normalizada de la portadora en dBW:

$$
P_c~[dBW] = 10~log(\frac{P_c~[W]}{1~W}) = 10~log(\frac{11250~W}{1~W})~~\to~~ P_c~[dBW] = 40,51~dBW
$$

### Parte B

Siendo la amplitud pico de una componente espectral:

$$
A_{mx} = m~.~A_c
$$

$$
A_{m1} = A_{m2} = A_{m3} = 0.8~.~150~V = 120~V
$$

### Parte C

Siendo la potencia normalizada de una banda lateral superior:

$$
P_{blsx} = \frac{(\frac{m_x~.~A_c}{2})^2}{2}
$$

Para $m = 80\%$:

$$
P_{bls1}~[W] = P_{bls2}~[W] = P_{bls3}~[W] = \frac{(\frac{0,8~.~150~V}{2})^2}{2} = 1800~W
$$

$$
P_{bls1}~[dBW] = P_{bls2}~[dBW] = P_{bls3}~[dBW] = 10~log(\frac{1800~W}{1~W}) = 32,56~dBW
$$

### Parte D

Siendo la potencia media normalizada:

$$
P_{NT}~[W] = P_{Nc}~[W] + P_{N1}~[W] + P_{N2}~[W] + P_{N3}~[W]
$$

Donde:

$$
P_{N1}~[W] = P_{N2}~[W] = P_{N3}~[W]~~~~~~~~~~~~~~~P_x~[W] = 2~.~P_{blsx}~[W]
$$

$$
P_{NT}~[W] = 11250~W + 6~.~1800~W
$$

$$
P_{NT} = 22050~W
$$

Siendo la potencia media con una cargo de $50~\Omega$:

$$
P_T = \frac{P_{NT}}{50~\Omega} = \frac{22050}{50~\Omega}~~\to~~P_T = 441~W
$$

### Parte E

Siendo la potencia pico de envolvente normalizada en Watts:

$$
PEP~[W] = \frac{A^2_c~.~(1 + m)^2}{2} = \frac{(150~V)^2~.~(1 + 0.8)^2}{2}~~\to~~PEP = 36,45~kW
$$

Siendo la potencia pico de envolvente normalizada en dBW:

$$
PEP~[dBW] = 10~log(\frac{PEP~[W]}{1~W})~~\to~~PEP~[dBW] = 45,61~dBW
$$
