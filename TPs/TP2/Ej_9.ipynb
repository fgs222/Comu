### Parte A

### Analisis tramo superior

Siendo:

$$
x(t) = A_m~.~cos(2\pi f_mt)~~~~~~~~~~~~con~~~f_m < 2~.~f_0
$$



Siendo la seal del oscilador 1:

$$
c_1(t) = A_{c1}~.~cos(2\pi f_0t)~~~~~~~~~~~~con~~~f_0 = \frac{W}{2}~~~A_{c1} = 2~V
$$



En A:

$$
V_1(t) = x(t)~.~c(t) = A_m~.~cos(2\pi f_mt)~.~A_{c1}~.~cos(2\pi f_0t)
$$

$$
V_1(t) = A_m~.~A_{c1}~.~cos(2\pi f_mt)~.~cos(2\pi f_0t)
$$

Siendo:

$$
cos(a)~.~cos(b) = \frac{1}{2}~.~cos(a + b) + \frac{1}{2}~.~cos(a - b)
$$

Por lo tanto:

$$
V_1(t) = \frac{A_m~.~A_{c1}}{2}~.~[cos(2\pi f_mt + 2\pi f_0t) + cos(2\pi f_mt - 2\pi f_0t)]
$$



Luego de pasar por el LPF1:

$$
V_1(t) = \frac{A_m~.~A_{c1}}{2}~.~cos(2\pi f_mt - 2\pi f_0t)
$$

Siendo la seal del oscilador 2:

$$
c_2(t) = A_{c2}~.~cos(2\pi f_1t)~~~~~~~~~~~~~~~~con~~~f_1 = f_c + f_0
$$



En C:

$$
V_3(t) = V_1(t)~.~c_2(t) = \frac{A_m~.~A_{c1}}{2}~.~cos(2\pi f_mt - 2\pi f_0t)~.~A_{c2}~.~cos(2\pi f_1t)
$$

$$
V_3(t) = \frac{A_m~.~A_{c1}~.~A_{c2}}{2}~.~cos(2\pi f_mt - 2\pi f_0t)~.~cos(2\pi f_1t)
$$

$$
V_3(t) = \frac{A_m~.~A_{c1}~.~A_{c2}}{4}~.~[cos(2\pi f_mt - 2\pi f_0t + 2\pi f_1t) + cos(2\pi f_mt - 2\pi f_0t - 2\pi f_1t)]
$$



### Analisis tramo inferior

Siendo la seal del oscilador 1 desfazada 90 grados:

$$
c_1(t) = A_{c1}~.~sin(2\pi f_0t)~~~~~~~~~~~~con~~~f_0 = \frac{W}{2}~~~A_{c1} = 2~V
$$



En B:

$$
V_2(t) = x(t)~.~c_1(t) = A_m~.~cos(2\pi f_mt)~.~A_{c1}~.~sin(2\pi f_0t)
$$

$$
V_2(t) = A_m~.~A_{c1}~.~cos(2\pi f_mt)~.~sin(2\pi f_0t)
$$

Siendo:

$$
cos(a)~.~sen(b) = \frac{1}{2}~.~sin(a + b) - \frac{1}{2}~.~sin(a - b)
$$

Por lo tanto:

$$
V_2(t) = \frac{A_m~.~A_{c1}}{2}~.~[sin(2\pi f_mt + 2\pi f_0t) - sin(2\pi f_mt - 2\pi f_0t)]
$$



Luego de pasar por el LPF2:

$$
V_2(t) = -\frac{A_m~.~A_{c1}}{2}~.~sin(2\pi f_mt - 2\pi f_0t)
$$

Siendo la seal del oscilador 2 desfazada 90 grados:

$$
c_1(t) = A_{c2}~.~sin(2\pi f_1t)~~~~~~~~~~~~~~~~con~~~f_1 = f_c + f_0
$$



En D:

$$
V_4(t) = V_2(t)~.~c_2(t) = -\frac{A_m~.~A_{c1}}{2}~.~sin(2\pi f_mt - 2\pi f_0t)~.~A_{c2}~.~sin(2\pi f_1t)
$$

$$
V_4(t) = -\frac{A_m~.~A_{c1}~.~A_{c2}}{2}~.~sin(2\pi f_mt - 2\pi f_0t)~.~sin(2\pi f_1t)
$$

Siendo:

$$
sen(a)~.~sen(b) = \frac{1}{2}~.~cos(a - b) - cos(a + b)
$$

$$
V_4(t) = -\frac{A_m~.~A_{c1}~.~A_{c2}}{2}~.~[cos(2\pi f_mt - 2\pi f_0t - 2\pi f_1t) - cos(2\pi f_mt - 2\pi f_0t + 2\pi f_1t)]
$$



### Analisis en el sumador

En el sumador:

$$
X_c(t) = V_3(t) + V_4(t)
$$
$$
= \frac{A_m~.~A_{c1}~.~A_{c2}}{4}~.~[cos(2\pi f_mt - 2\pi f_0t + 2\pi f_1t) + cos(2\pi f_mt - 2\pi f_0t - 2\pi f_1t)]
$$
$$
- \frac{A_m~.~A_{c1}~.~A_{c2}}{2}~.~[cos(2\pi f_mt - 2\pi f_0t - 2\pi f_1t) - cos(2\pi f_mt - 2\pi f_0t + 2\pi f_1t)]
$$

$$
X_c(t) = \frac{A_m~.~A_{c1}~.~A_{c2}}{4}~.~[cos(2\pi f_mt - 2\pi f_0t + 2\pi f_1t) + cos(2\pi f_mt - 2\pi f_0t - 2\pi f_1t)
$$
$$
- cos(2\pi f_mt - 2\pi f_0t - 2\pi f_1t) + cos(2\pi f_mt - 2\pi f_0t + 2\pi f_1t)]
$$

$$
X_c(t) = \frac{A_m~.~A_{c1}~.~A_{c2}}{4}~.~[cos(2\pi f_mt - 2\pi f_0t + 2\pi f_1t) + cos(2\pi f_mt - 2\pi f_0t + 2\pi f_1t)]
$$

$$
X_c(t) = \frac{A_m~.~A_{c1}~.~A_{c2}}{2}~.~cos(2\pi f_mt - 2\pi f_0t + 2\pi f_1t)
$$

Remplazando por los valores del enunciado:

$$
X_c(t) = \frac{A_m~.~2~V~.~A_{c2}}{2}~.~cos(2\pi f_mt - 2\pi \frac{W}{2}t + 2\pi (f_c + f_0)t)
$$

$$
X_c(t) = (A_m~.~A_{c2})~V.~cos(2\pi f_mt - 2\pi\frac{W}{2}t + 2\pi (f_c + \frac{W}{2})t)
$$

$$
X_c(t) = A_m~.~A_{c2}~V.~cos(2\pi f_mt + 2\pi f_ct)
$$

Se observa que este modulador permite obtener la banda lateral superior.
