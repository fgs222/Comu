# Ejercicio 7

### Parte A

Siendo:

$$
x(t) = \frac{1}{2}~.~A_c~.~A_m~.~cos[(\omega_c - \omega_m)~.~t]
$$

Siendo:

$$
c(t) = A_c~.~cos(\omega_ct + \phi)
$$

Siendo:

$$
s(t) = x(t)~.~c(t) = \frac{1}{2}~.~A_c~.~A_m~.~cos[(\omega_c - \omega_m)~.~t]~.~A_c~.~cos(\omega_ct + \phi)
$$

$$
s(t) = \frac{A^2_c~.~A_m}{2}~.~cos[(\omega_c - \omega_m)~.~t]~.~cos(\omega_ct + \phi)
$$

Para que el oscilador local se encuentra sincronizado en frequencia y fase en el transmisor y receptor:

$$
\phi = 0
$$

$$
s(t) = \frac{A^2_c~.~A_m}{2}~.~cos[(\omega_c - \omega_m)~.~t]~.~cos(\omega_ct)
$$

Siendo:

$$
cos(a)~.~cos(b) = \frac{1}{2}~.~cos(a + b) + \frac{1}{2}~.~cos(a - b)
$$

Donde:

$$
a = (\omega_c - \omega_m)~.~t~~~~~~~~b = \omega_ct
$$

Aplicando en $s(t)$:

$$
s(t) = \frac{A^2_c~.~A_m}{2}~.~(\frac{1}{2}~.~cos[\omega_ct - \omega_mt + \omega_ct] + \frac{1}{2}~.~cos[\omega_ct - \omega_mt - \omega_ct])
$$

$$
s(t) = \frac{A^2_c~.~A_m}{4}~.~(cos[2\omega_ct - \omega_mt] + ~cos[-\omega_mt])
$$

Como la funcion coseno es par:

$$
s(t) = \frac{A^2_c~.~A_m}{4}~.~(cos[2\omega_ct - \omega_mt] + ~cos[\omega_mt])
$$

Se determina la frecuencia de corte del LPF para obtener el mensaje y se obtiene:

$$
y(t) = \frac{A^2_c~.~A_m}{4}~.~cos(\omega_mt)
$$

Podemos observar que el receptor funciona para obtener una seal de banda lateral inferior teniendo en cuenta las caracteristicas y sintonizando el receptor adecuadamente.

### Parte B

Siendo la seal demodulada:

$$
v_o(t) = \frac{A_c~.~A_0}{4}~.~[H(f_c - f_m) + H(f_c + f_m)]~.~m(t)
$$

Es necesario que la respuesta al impulso del filtro en frecuencia H(f) sea igual a uno para todo valor de frecuencia menor a la frecuencia maxima del ancho de banda y cero para cualquier otro valor de frecuencia.

### Parte C y D

Debido a que el receptor sincronico es coherente, tanto el transmisor como el receptor deben estar sincronizados en frecuencia y en fase. En caso de que esto no se cumpla, la seal que se obtiene en el receptor no va a ser igual al mensaje transmitido por el transmisor.
